```python
# 리스트일때

num_list=["one", "two", "three"]

for i in num_list:
    print(i)

# 튜플
a=[(1, 2), (3, 4), (5, 6)]
for j in a:
    print(j)


# 언패킹
a=[(1, 2), (3, 4), (5, 6)]
for first, second in a:
    print(first, second)  #변수 first, second 가 튜플 안의 요소를 각각 꺼내오게 하는 거임



# 튜플은 괄호 없이도 가능하기 때문에
a, b= 1, 2
[a, b]=[1, 2]
[a, b]= (1, 2)


```

```python
marks=[90, 25, 67, 45, 80]

number=0
for mark in marks:
    number+=1
    if mark>60:
        print(f"{number}번째 학생은 통과입니다.")
    else:
        print(f"{number}번째 학생은 불합격입니다.")
```

```python
# for문과 continue문

marks=[90, 25, 67, 45, 80]

number=0

for mark in marks:
    number+=1
    if mark<60:
        continue
    print(f"{number} 번째 학생은 합격입니다.")



```

```python
# for문과 range 함수

a=range(10)
print(a)

a= range(10)
print(list(a))

a= range(1, 10)    # *************************************8마지막 인덱스에서 -1 한게 마지막 값임***************************************
print(list(a))

result=0
for number in range(1, 11):
    result+=number
    print(result)   # 더하고 있는 과정 출력
print(result)     #결과 출력



```

```python
marks=[90, 25, 67, 45, 80]

len(marks)

for number in range(len(marks)):
    if marks[number]<60:
        continue
    print(f"{number+1} 번째 학생은 합격입니다")


    
```

```python
# 중첩 for문
# for와 range를 이용한 구구단 

for i in range(2, 10):
    for j in range(1, 10):
        print(i*j, end=" ")
    print("")

```

```python
# 출력 결과뭉
# 2x1=2
# 2x2= 4



for i in range(2, 10):
    for j in range(1, 10):
        result=i*j
        print(f"{i}x{j}={result}")    #result 대신에 바로 i*j 써도 무방함
    print("")




```

```python
# 리스트 컴프리헨션 사용하기
a= [1, 2, 3, 4]
result= []

for num in a:
    result.append(num*3)
print(result)

```

```python
# for문과 break문

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
#enumerate()함수

fruits= ["사과", "바나나", "오렌지"]

for i, fruit in enumerate(fruits):
    print(f"{i} 번째:{fruit}")


#zip 함수로 여러 리스트 함께 순회하기
names= ["홍길동", "유재석", "강호동"]
scores= [60, 70, 80]

for name, score in zip(names, scores):
    print(f"{name}은 {score} 점")

# 3개 이상의 리스트-zip
names= ["홍길동", "유재석", "강호동"]
korean= [60, 70, 80]
english=[80,90,70]

for name, k, e in zip(names, korean, english):
    print(f"{name}의 국어 점수는 {k}점, 영어점수는 {e}점")

```
