```python
treehit= 0

while treehit <10:
    treehit= treehit+1
    print(f"나무를 {treehit}번 찍었습니다")
    if treehit==10:
        print("나무가 넘어갑니다")
```

```python
number=0
prompt="""
1. Add
2. Del
3. List
4. Quit
"""

while number != 4:
    print(prompt)
    number= int(input("1~4까지 숫자를 입력해주세요"))
```

```python
#  while 문 강제로 빠져나가기
커피= 5
돈= 300


while 돈:
    print("돈을 받았으니 커피를 줍니다")
    커피-=1
    print(f"남은커피는 {커피}잔 입니다")
    if 커피==0:
        print("커피 다 팖, 판매 중지")
        # 여기까지하면 무한루프에 빠짐
        break
    

```
