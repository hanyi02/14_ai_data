```python
# 튜플은 어떻게 만드는가

a= () # 빈튜플
b=(1,) # 요소가 하나일 때 콤마를 찍어줘야 함----> 튜플은 값이 아니라 콤마로 만들어지기 때문
c=(1, 2, 3)
d= 1, 2, 3
e= ("a", "b", ("ab", "cd"))
f= 1,
g= 1
h=(1) ####################### 괄호 안에 숫자 하나면 튜플 아니고 그냥 int임


print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))
print(type(h))
```

```python
#  튜플 요소 값 삭제
a=(1, 2, "a", "b")
del a[0] # 튜플 값은 삭제가 안 됨

print(a)
```

```python
#  튜플 값 변경
a=(1, 2, "a", "b")
a[0]="c" # 튜플 값은 삭제도 안 됨

print(a)
```

```python
# 튜플 연산
#---------> 위에 값 변경이랑 삭제가 안 되는데 연산은 되는 이유가 튜플이 객체참조형이기 때문임 /// 포인터 개념이랑 다름
a=(1, 2)
a= a*2
print(a)

a=10
print(id(a))
a=20
print(id(a))
print(id(10))

#-----------> 리스트는 mutable 객체이고
# 튜플은 immutable(불변) 객체이다
```

```python
# 튜플에 인덱싱

a= (1, 2, 3, 4)
print(a[0])

# 튜플에 슬라이싱
a=(1, 2, 3, 4)
print(a[:2])

# 튜플 + 튜플
a=(1, 2)
b=(3, 4)

c= a+b #새롭게 생성이 된 거
print(c)

# 튜플 * 튜플
a=(1, 2)
b=a*3
print(b)

# 튜플 길이 구하기
a=(1, 2, 3, 4, )
len(a)
```
