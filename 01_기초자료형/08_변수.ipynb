```python
# 변수는 어떻게 만드는가?
# 변수 이름= 변수에 저장할 값

# 변수명 규칙
# 1. 영문자, 숫자, 언더스코어(_) 만 사용가능
# 2. 숫자로는 시작할 수 없음
# 3. 예약어(명령어)는 변수로 사용 불가함
# 4. 대소문자 구분됨


name="홍길동"

user_name= "홍길동"

# 컨벤션 
# user_name 스네이크 방식
# userName 카멜
# UserName 파스칼--> 클래스는 무조건 대문자로 시작, 함수명은 스네이크로
 
```

```python
a= [1, 2, 3] 
print(id(a))
b= [1, 2, 3]
print(id(b))


a.append(4)
print(id(a))  #인덱스는 mutable 객체이므로 값이 변할 수 있음->>> 주소가 같음/ 근데 뮤터블이라 키로 사용은 안 됨

b=a
print(id(b))
```

```python
a= [1, 2, 3]
b=a[:] #복제

print(id(a))
print(id(b)) #복제긴한데 주소는 다름

```

```python
# 변수를 만드는 여러가지 방법

a, b=["python", "like"] #양변 개수는 맞춰줘야 함
print(a)  # 언팩
print(b)

print(type(a))


```

```python
(a, b)="python", "like" # 튜플로 생성, 좌변에 괄호 없어도 튜플임
print(a)  # 언팩
print(b)

print(type(a))
```

```python
a=b="python"
print(a)
print(b)
print(type(a))


```
