```python
# 문자열 포매팅
#  1. 숫자 바로 대입

print("I eat %d apples" %3) #d에 대입-> decimal

# 2. 문자열 바로 대입
print("I eat %s apples" % "five") #s에 대입-> string

# 3. 숫자 값을 나타내는 변수로 대입
number= 3
print("I eat %d apples" % number) #d에 대입-> 변수 안에 decimal 저장이라 d 사용

# 4. 2개 이상의 값 대입
number= 10
day= "three"

print("I ate %d apples, so i was sick for %s days." % (number, day) ) 
```

### 문자열 포맷 코드
- %s: 문자열(string)
- %c: 문자 1개(character)
- %d: 정수(integal)
- %f: 실수(float)
- %o: 8진수
- %x: 16진수
- %%: 문자 % 자체 출력

```python
# format 함수 사용한 포매팅
# 1. 숫자 바로 대입하기
print("I eat {0} apples".format(3))


# 2. 문자열 바로 대입하기
print("I eat {0} apples".format("five"))

# 3. 2개 이상의 값 넣기
number= 10
day="three"
print("I ate {0} apples, so i was sick for {1} days.".format(number, day))
```

```python
# f문자열 포매팅
name= "홍길동"
age= 30
print(f"나의 이름은 {name}입니다. 나이는 {age}입니다. ")
```

```python
# 문자열 관련 함수들

# 1. 문자 개수 세기 -> count
a="hobby"
print(a.count("b")) #왜 a 뒤에 콤마가 붙는지 이해할 것--> 원래 len이나 type 같은 함수는 내장함수라서 그냥 사용가능함. 근데 a라는 객체가 문자열일때만 사용가능한 count는 매서드(객체에 속한 함수)라서 콤마를 통해 불러오는 것.

# 2. 위치 알려주기-> find
print(a.find("b")) # 이것도 0부터 시작함.
print(a.find("k")) # 없으면 결과 -1로 나옴

# 3. 위치 알려주기
print(a.index("b")) # 위치이기 때문에 숫자로 나옴

# 4. 문자열 삽입 -> join: 문자 사이사이에 ,삽입
print(",".join("abcd"))

# 5. 소문자를 대문자로 만들기-> upper
a="hello"
print(a.upper())
#### 한글자만 대문자로 만들려면 결합해야함
print(a[0].upper()+a[1:]) #-----> Hello 출력

# 6. 대문자를 소문자로 만들기-> lower
a="hello"
print(a.lower())

# 7. 공백지우기 -> strip
a=" hi  "
print(a.strip())

# 8. 왼쪽 공백지우기 -> lstrip
# 9. 오른쪽 공백지우기 -> rstrip
print("왼쪽", a.lstrip()) # ,사용이 공백 하나 추가해줌
print("왼쪽"+ a.lstrip()) # +사용시 공백 없이 붙여줌
print("오른쪽", a.rstrip())


# 10. 문자열 바꾸기-> replace
a="Life is too short"
a.replace("Life", "Your leg")

# 11. 문자열 나누기-> split
a="Life is too short"
print(a.split()) #아무것도 안 적으면 공백을 기준으로 나누어줌, ***********리스트로 출력됨*************
print(a.split("i")) # i 기준으로 나누어지기 때문에 i는 없어지고 리스트 형태로 출력됨
```
