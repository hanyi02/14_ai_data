```python
# bool 자료형 사용법
a= True
b= False

print(a)
print(type(a)) #<class 'bool'>

print(a==b)
print(0)
print(1)

print(1, bool(1))
print("Python", bool("Python"))
print("Python", bool("llll"))  # True ---> 문자열에서는 빈칸만 넣는 경우 빼고는 다 T 나옴
print("[]", bool([])) #하면 요소 없어서 F
print("[]", bool([1])) #하면 요소 있어서 T


# if의 logical test에 bool 사용 (참고)
if 1>3:
	print("참") 
else:
	print("거짓")

```

```python
# 논리연산자 (and, or, not)

# 1. and
print(True and True) #############둘 다 T일 때만 T 반환
print(True and False)
print(False and False)
print(1 and 1) ################## 둘 다 1일 때만 1 반환
print(1 and 0)
print(0 and 0)


# 2. or
print(True or True) #############하나이상 T일 때 T 반환
print(True or False) #T
print(False or False) 
print(1 or 1) ################## 하나 이상 1일 때 1 반환
print(1 or 0)
print(0 or 0)

# 3. not
print(not True) #F
print(not []) #T
```
