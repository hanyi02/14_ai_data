```python
# 딕셔너리 쌍 추가
a={1: "a"}
a[2]="b"
print(a) #{1: 'a', 2: 'b'} ------> 리스트 값 추가 할 때는 인덱스 값 기반으로 추가했는데, 딕셔너리는 키 값 기준으로 값 추가함 **********인덱스 번호 아님**************

a["name"]="ken" # 딕셔너리 추가로 int랑 string은 가능함
print(a)

a[(1, 2)]="hobby" #튜플도 가능함
print(a)

a[[1, 4]]="job"
print(a) # 리스트는 추가 안 됨###########################################
```

```python
# 딕셔너리 요소 삭제하기

print(a)

del a[1] ######################## 이때까지는 인덱스로 삭제했는데 딕셔너리는 키 값으로 삭제함
print(a)

del a["name"] ######## 왜 안 되냐면 위에서 키 값 삭제함으로써 참조 주소 바꼈는데 또 삭제하려니 에러남
print(a)
```

```python
# 딕셔너리 관련 함수

# 1. key 리스트 만들기: keys

profile={
	"이름": "홍길동", 
	"나이":25,
	"번호":"011"
	}

print(profile.keys())  # dict_keys(['이름', '나이', '번호'])


# 2. vlaue 리스트 만들기: values

profile={
	"이름": "홍길동", 
	"나이":25,
	"번호":"011"
	}

print(profile.values()) # dict_values(['홍길동', 25, '011'])


# 3. key: value 쌍 얻기: items

profile={
	"이름": "홍길동", 
	"나이":25,
	"번호":"011"
	}

print(profile.items()) #dict_items([('이름', '홍길동'), ('나이', 25), ('번호', '011')])---> ([(    )])  구조 *****************************


# 4. key, value 쌍 모두 지우기-> clear
profile={
	"이름": "홍길동", 
	"나이":25,
	"번호":"011"
	}

print(profile.clear()) #None


# 5. key로 value 값 얻기: get()

profile={
	"이름": "홍길동", 
	"나이":25,
	"번호":"011"
	}

print(profile.get("이름")) #profile["이름"]

# 6. 해당 key가 딕셔너리 안에 있는지 조사하기: in

profile={
	"이름": "홍길동", 
	"나이":25,
	"번호":"011"
	}

print("이름" in profile)   #True




# 7. key로 value 얻기: pop


profile={
	"이름": "홍길동", 
	"나이":25,
	"번호":"011"
	}
temp= profile.pop("이름")
print(temp)  #홍길동
print(profile) #{'나이': 25, '번호': '011'}
```
