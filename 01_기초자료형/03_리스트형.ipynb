```python
# 리스트는 어떻게 만드는가?
# 형식

a=[]
b=[1, 2, 3]
print(b[1]) #문자열이랑 똑같이 0부터 시작함
c=["Life", "is", "too", "short"]
d= [1, 2, "Life", "is"]
e= [1, 2, ["Life", "is"]]
# print(e.split(",")) -> *********** 리스트는 원래 나뉘어 져 있어서 split이 없음. 다시 나누려면 조건문 사용해야함.

print(a) #--> []
print(type(a)) 
print(len(a)) #------> *****************문자열 인덱싱과 다르게 len은 1부터 시작함***************
print(len(b))
print(len(c))
print(len(d))
print(len(e))

```

```python
# 리스트의 인덱싱

a= ["Life", "is", "too", "short"]
print(a[2])
print(a[-1])
print(a[1:3]) #------> 슬라이싱 사용 시 리스트로 나옴

print(a[0][3]) # "Life"의 e만 출력//// a[0][-1]도 같은 의미
```

```python
a= [1, 2, 3, ["a", "b", "c"]]
print(a[-1])
print(a[-1][1])
print(a.index(2)) #--> 숫자 2 위치 출력
```

```python
# 리스트의 슬라이싱
a=[1, 2, 3, 4, 5]
print(a[0:1+1])
print(a[-2:])
print(a[:-1])
```

```python
# 리스트 연산
# 1. 리스트 더하기

a=[1, 2, 3]
b=[4, 5, 6]

print(a+b) #---> [1, 2, 3, 4, 5, 6]
result = [a[i] + b[i] for i in range(len(a))] #각 자리 더하기--> [5, 7, 9]
print(result)

# 2. 리스트 곱하기(반복)
a=[1, 2, 3]
b=[4, 5, 6]
print(a*3) #-----> 진짜 단순 반복


```

```python
#  리스트의 수정과 삭제
# 1. 리스트 수정

a=[1, 2, 3]
a[2]=4
print(a)

# 2. 리스트 요소 삭제
a=[1, 2, 3]
del a[1]  #del->delete 단축어
print(a) #--> [1, 3]

a=[1, 2, 3, 4, 5]
del a[2:]
print(a)

a = [1, 2, 3]
a.remove(1)
print(a)
```

```python
# 리스트 관련 함수
# 1. 리스트에 요소 추가하기 append

a= [1, 2, 3]
a.append(4)
print(a) #----> [1, 2, 3, 4]

a = [1, 2, 3]
a.insert(1, 4)
print(a) #------> [1, 4, 2, 3 ***** 1자리에 4추가 ******


a.append([5, 6])
print(a) #----> [1, 4, 2, 3, [5, 6]]

del a[4]
print(a) #---> [1, 4, 2, 3]
```

```python
# 2. 리스트 정렬 sort

# 오름차순
a= [1, 4, 2, 3]
a.sort()
print(a) # [1, 2, 3, 4]

# 내림차순
a= [1, 4, 2, 3]
a.sort(reverse=1) # true가 기본값=1
print(a)


```

```python
# 3. list 뒤집기-> reverse

a=[1, 2, 3, 4]
a.reverse()
print(a)

a=["a", "d", "c"]
a.sort()
print(a)
a.reverse()
print(a)
```

```python
# 4. 인덱스 반환-> index

a= [5, 4, 3, 2, 1]
print(a.index(3))
print(a.index(2))
```

```python
# 5. 요소 삽입

a=[1, 2, 3]
a.insert(0, 4)
print(a)
```

```python
# 6. 리스트 요소 제거-> remove
a=[1, 2, 3, 1, 2, 3]
a.remove(3) #-----> 첫번째 값만 지워짐
print(a)


#print(a.insert(0,4)) 하면 None 뜸--> insert()는 리스트를 변경만 하고, 값을 반환하지 않음
```

```python
# 7. 리스트 요소 꺼내기
a=[1, 2, 3]
result1= a.pop() #----> 맨 뒤에 인수값 반환
print(a)
print(result1)
```

```python
# 8. 요소의 개수 세기
a=[1, 2, 3, 4, 1]
a.count(1)
```

```python
# 9. 리스트 확장-> extend
a= [1, 2, 3]
a.extend([4, 5])
print(a) #[1, 2, 3, 4, 5]--> []로 안 묶임
```
