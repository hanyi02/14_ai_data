```python
# ctrl+ / -> # 
# 큰 따옴표 둘러싸기
"life is too short, you need Python"

# 작은따옴표 둘러싸기
'python is fun'

# 큰 따옴표 3개 연속으로 둘러싸기-> 줄 바꿔도 에러 안 생김(따옴표 한 개면 줄 바꿈 시 에러남)
"""life is too short, 
you need Python"""

# 작은 따옴표 3개
''' life is too short,
 you need Python'''
```
