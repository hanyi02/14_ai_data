```python
#  집합 자료형은 어떻게 만드는가?

a=set([1, 2, 3])
print(type(a)) #<class 'set'>
print(a) #{1, 2, 3}


b=set("hello")
print(b)

c={1, 2, 3}
print(type(c))
print(c)

d={"a", "b", "c"}
print(d) #{'c', 'a', 'b'}---> 순서 없음

```

```python
# 1. 교집합
s1= set([1, 2, 3, 4, 5, 6])
s2= set([4, 5, 6, 7, 8, 9])

print(s1.intersection(s2)) #{4, 5, 6} -----------------------> 머신러닝 시 중복부분 제거 시 유용하게 쓰임

print(s1 & s2) #{4, 5, 6}


# 2. 합집합
s1= set([1, 2, 3, 4, 5, 6])
s2= set([4, 5, 6, 7, 8, 9])

print(s1.union(s2)) #{1, 2, 3, 4, 5, 6, 7, 8, 9}

print(s1 | s2)


# 3. 차집합
s1= set([1, 2, 3, 4, 5, 6])
s2= set([4, 5, 6, 7, 8, 9])

print(s1.difference(s2)) #{1, 2, 3}
print(s2.difference(s1)) #{8, 9, 7}
print(s1-s2) #{1, 2, 3}
print(s2-s1) #{8, 9, 7}


# 4. 값 추가하기
s1=set([1, 2, 3])
s2=set([4, 5, 6])
s1.add(4) 
print(s1) #{1, 2, 3, 4}

# 5. 여러개 추가하기: update
s1=set([1, 2, 3])
s1.update([4, 5, 6])
print(s1) #{1, 2, 3, 4, 5, 6} --> 리스트가 아니라 집합으로 나옴


# 6. 특정 값 제거하기: remove
s1=set([1, 2, 3])
s1.remove(2)
print(s1) #{1, 3}


# 7. 모든 값 제거하기
s1=set([1, 2, 3])
s1.clear()
print(s1) #set()
```
