```python
# 파일 생성하기

f=open("새파일.txt", "w")
f.close()
```

```python
f = open('새파일2.txt', 'w', encoding='utf-8')
for i in range(1, 11):
    data = f'{i}번째 줄입니다.\n'
    f.write(data)
f.close()

# 안 깨진 문자 보려면 utf8 해야 인코딩/ 디코딩 오류 없음

```

```python
f = open('./새파일2.txt', 'r', encoding='utf-8')
line = f.readline()
print(line)
f.close
```

```python
f = open('./새파일2.txt', 'r', encoding='utf-8')
while True:
    line = f.readline()
    if not line: break
    print(line)
f.close
```

```python
# readlines()
f = open('./새파일2.txt', 'r', encoding='utf-8')
lines = f.readlines()
print(lines)
f.close()
```

```python
# 새로운 내용 추가하기
f = open('./새파일2.txt', 'a', encoding='utf-8')
for i in range(11, 21):
    data = f'{i}번째 줄입니다.\n'
    f.write(data)
f.close()
```

```python
# with 문과 함께 사용하기
with open('./새파일.txt', 'w', encoding='utf-8') as f:
    f.write('Life is too short, you need python')
```
