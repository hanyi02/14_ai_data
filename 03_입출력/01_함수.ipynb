```python
# 함수의 구조

# def 함수 이름(매개변수 1, 매개변수 2, ...):
    # 수행할 문장 1
    # 수행할 문장 2
    #  ...

def add(a, b):
    return a+b


print(add(1, 2))

```

```python
# 1. 일반적인 함수
# 입력값이 있고, 반환값이 있다.


def add(a, b):
    result=a+b
    return result

sum= add(3, 4)
print(sum)


```

```python
# 2. 입력값이 없는 함수
# 입력값 x, 반환값 o

def say():
    return "hi"

a= say()
print(a)

```

```python
# 3. 반환값이 없는 함수
# 입력값은 있고, 반환값은 없음


def add(a, b):
    print(f"{a},{b}의 합은 {a+b}입니다.")


sum=add(3, 4)
print(sum)
```

```python
# 4. 입력값, 반환값 둘 다 없는 함수


def say():
    print("hi")



a= say()
print(a)
```
