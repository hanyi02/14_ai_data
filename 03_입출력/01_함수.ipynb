```python
# 6. 입력값이 몇 개가 될지 모를때는 어떻게 하는가?


def add_man(*args):
    result=0
    print(type(args))   #튜플 나옴
    print(args)
    for i in args: #튜플은 in애 들어갈 수 있음
        result+=i
    return result

r= add_man(1, 2, 3, 4, 5)
print(r)

```

```python
def add_mul(choice, *args):
    if choice== "add":
        result=0
        for i in args:
            result+=i
    elif choice=="mul":
        for i in args:
            result *= i

    return result


result= add_mul("add", 1, 2, 3, 4, 5)
print(result)
```

```python
# 7. 키워드 매개변수, kwargs
def print_kwargs(**kwargs):
    print(kwargs)

print_kwargs(a=1) #{'a': 1}---> 딕셔너리로 나옴


print_kwargs(name="foo", age=20, job="student") #{'name': 'foo', 'age': 20, 'job': 'student'}



def create_profile(**kwargs):
    print("===== 프로필 정보 ======")
    for key, value in kwargs.items():
        print(f" {key}: {value}")


create_profile(이름= "김철수", 취미="축구", 나이= 25)
```

```python
# lambda 식 예약어

# 함수_이름= lambda 매개변수1, 매개변수2, ... : 수행할 문장




def add(a, b):
    return a+b

add= lambda a, b: a*b
```

```python
# 함수의 독스트링(Docstring)
def add(a, b):
    """
    두 숫자를 더하는 함수

    Parameters:
        a (int, float): 첫 번째 숫자
        b (int, float): 두 번째 숫자

    Returns:
        int, float: 두 숫자의 합
    """
    return a + b

print(add.__doc__)
```
