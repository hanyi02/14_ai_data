```python
#  클래스와 def의 차이
# -> def는 기능하나, 클래스는 안에 def 여러개 만들 수 있음




# 계산기
class Calculator:
    def __init__(self):
        self.result = 0
    
    def add(self, num):
        self.result += num
        return self.result
    
    def sub(self, num):
        self.result -= num
        return self.result
    
    def mul(self, num):
        self.result *= num
        return self.result
    
    def div(self, num):
        self.result /= num
        return self.result
    
    def reset(self):
        self.result = 0
        return self.result
    
cal1 = Calculator()
print(cal1.add(5))
print(cal1.add(10))
print(cal1.add(10))
print(cal1.reset(),'\n')

print(cal1.add(10))
print(cal1.sub(5))
print(cal1.mul(4))
print(cal1.div(2))
print(cal1.reset())
```

```python
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second
    
    def add(self):
        result = self.first + self.second
        return result
    
    def sub(self):
        result = self.first - self.second
        return result
    
    def mul(self):
        result = self.first * self.second
        return result
    
    def div(self):
        result = self.first / self.second
        return result
    
cal4 = FourCal()
cal4.setdata(10,5)

print(cal4.add())
print(cal4.sub())
print(cal4.mul())
print(cal4.div())
```

```python
# __init__ 메서드도 첫 번째 매개변수 self에 생성되는 객체가 자동으로 전달
# __init__ 메서드가 호출되면 setdata 메서드를 호출했을 때와 마찬가지로 first와 second라는 객체변수가 생성

class FourCal:
    def __init__(self, first, second):
        self.first = first
        self.second = second
    
    def add(self):
        result = self.first + self.second
        return result
    
cal5 = FourCal(10, 20)
print(cal5.first)
print(cal5.second)

print(cal5.add())
```
